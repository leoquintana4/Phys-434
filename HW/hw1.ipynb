```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Generaate 100k data points for a gaussian distribution with mean 3
mu = 3
std = .01
data = stats.norm.rvs(loc = mu, scale = std, size = 100000)
size = data.shape[0]
# Show number of data points
print("Number of sample data points", size)

# Plot histogram with 50 bins and x-axis range 2.95 to 3.05 and add axis labels and title
plt.hist(data, bins =50, range=(2.95,3.05))
plt.xlabel("Values")
plt.ylabel("Frequency of Points")
plt.title("Histogram of Gaussian Distribution") 

#Create Gaussian PDF with normalization to 100k data points
# Create equal spaced x avlues to plot the PDF
x = np.linspace(2.95, 3.05, 2000)
y = stats.norm.pdf(x, loc = mu, scale = std) * size * [(3.05-2.95)/50]

#plot the PDF on top of the histogram
plt.plot(x, y)

# Add y-axis in log scale
plt.yscale("log")
plt.show()
```

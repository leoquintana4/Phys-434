### 1. Estimating π 
### Let's estimate the numerical value of π


#### a. Generate N=1000 random points (x,y) within a square by requiring the range of x and y are both within [0, 1].

#### b. Visualize the result with red dots when data points are inside the unit circle (otherwise blue dots). In the figure, draw the unit circle as boundar line.

#### c. Count number of points inside the unit circle, and estimate the value and uncertainty of π


#### d. Repeat steps abobe with N=10K and 1000K. Discuss the results in terms of numerical consistency with known  π value, and scaling rule of the uncertainty.

```python
import numpy as np
import scipy.stats as stats 
import matplotlib.pyplot as plt
# Create random points in square for both x and y
np.random.seed(42)
N = 1000
x = np.random.uniform(0,1, N)
y = np.random.uniform(0,1, N)

# Determine which points are inside and outside the unit circle
inside = x**2 + y**2 < 1
outside = x**2 + y**2 >= 1

# Plot the points and color them red if inside the circle and blue if outside
fig, ax = plt.subplots()
ax.scatter(x[inside], y[inside], color = 'red', s = 1, label = "inside circle")
ax.scatter(x[outside], y[outside], color = 'blue', s = 1, label = "outside circle")
# Draw a unit circle to visualize the boundry
circle = plt.Circle((0, 0), radius=1, edgecolor='black', facecolor='None')
ax.add_patch(circle)
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
plt.legend()

# Get the sum of points inside and outside the unit circle
inside_count = np.sum(inside)
outside_count = np.sum(outside)


print(f'Number of points inside unit circle" {inside_count})')
print(f'Number of points outside unit circle" {outside_count})')

# Calculate the pi estimate
pi_estimate = (inside_count / N) * 4
# Calculate the uncertainty using binomial std
pi_uncertainty = np.sqrt(inside_count * (1-inside_count / N)) / (N*4)
print(f'Estimated value of pi: {pi_estimate} +/- {pi_uncertainty:.4f}')
```

### 2. Parabola distribution: 
#### Generate a random distribution between x= -1 and x= +1 from the distribution function f(x)=x**2

#### a. Using Rejection method. Generatr a random variable x_i uniformly distributed between -1 and +1, and a second random ravariable y_i unformly distributed between 0 and f(1), i.e. allowed range of the distribution function. Keep x_i if y_i < f(x_i). Generate a sample with N=10K data points.

#### b. Using Inverse transformation method. Caculate the cumulative function y=F(x) and derive the inverse transformation function x=F−1(y). Generate a random uniform distribution  y_i between [-1, 1], and obtain the target distribution x_i with N=10K point.

#### c. Compare consistency by drawing two simulated distrubtions and the target distribution function f(x) in the same plot. Comment on which menthod is more effective in terms of total number of random number reuqired for MC simulations.



```python
# Question 2a:
import numpy as np
np.random.seed(42)
N = 10000 # number of samples
sample = [] # create empty list to store samples
for i in range(N): # loop over 
    x = np.random.uniform(-1, 1) # generator x_i
    y = np.random.uniform(0, 1) # generator y_i
    if y < x**2: # check if y_i is less than f(x_i) 
        sample.append(x) # reject if y_i > f(x_i)
        
sample = np.array(sample) # convert list to array
        
print(f'Samples values that have the y_i < f(x_i) condition true: {sample}')

# Question 2b:
np.random.seed(42)
N = 10000 # number of samples
y2 =np.random.uniform(-1,1, N) # generate uniform distribution between -1 and 1    
x_cdf = np.cbrt(3*y2) # apply inverse transformation function
print(f'Samples values from inverse transformation method: {y2}')

 # Plot both methods together 
ax, fig = plt.subplots()
plt.hist(3/2 * sample, bins = 40, density = True, alpha = 0.5, label = 'Rejection method', color = 'blue')
plt.hist(x_cdf, bins = 40, density = True, alpha = 0.5, label = 'Inverse transformation method', color = 'red')

# Plot the target distribution f(x) = x**2
x = np.linspace(-1, 1, 1000)
plt.plot(3/2 * x, x**2, label = 'Target Distribution', color = 'black')

# Question 2c:
# In our case the both methods give similar results, however 
# in my opinion the inverse transformation method is slightly 
# smoother and is much easier to implement here where target function isn't too complex
# although I believe both methods are use in their own ways depending on the problem at hand.
```

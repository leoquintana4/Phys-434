```python
# Create the sample
np.random.seed(42)
x = np.random.uniform(0, 1, 10000)
u = np.exp(-x)
u_values = np.linspace(0, 1, 100)
def g(u):
    result = 1/u
    return result

# Plot for uniform distribution
plt.figure(figsize=(10, 6))
plt.hist(x, bins=30, density=True, alpha=0.6, color='b', label='Histogram of f(x)')

# Plot the inversed function
plt.figure(figsize=(10, 6))
plt.hist(u, bins=30, density=True, alpha=0.6, color='g', label='Histogram of u(x)')


# Plot for g(u)
plt.figure(figsize=(10, 6))
plt.plot(u_values, g(u_values), alpha=0.6, color='r', label='Plot of g(u)')




```

```python
### Question 3

np.random.seed(4)
N = 10
p = 0.02
M = 10000
# create the binomial distribution
binomial_samples = np.random.binomial(N, p, M)
# create the normal distribution
normal_samples = np.random.normal(N*p, np.sqrt(N*p*(1-p)), M)
# create the poisson distribution
poisson_samples = np.random.poisson(N*p, M)

plt.figure(figsize=(10, 6))
plt.hist(binomial_samples, bins=25, density=True, alpha=0.6, color='b', label='Binomial Distribution') 
plt.hist(normal_samples, bins=25, density=True, alpha=0.6, color='r', label='Normal Distribution')
plt.hist(poisson_samples, bins=25, density=True, alpha=0.6, color='g', label='Poisson Distribution')

# Create table of means, sigma, skewness, and kurtosis with pandas
data = {
    'Distribution': ['Binomial', 'Normal', 'Poisson'],
    'Mean': [np.mean(binomial_samples), np.mean(normal_samples), np.mean(poisson_samples)],
    'Sigma': [np.std(binomial_samples), np.std(normal_samples), np.std(poisson_samples)],
    'Skewness': [stats.skew(binomial_samples), stats.skew(normal_samples), stats.skew(poisson_samples)],
    'Kurtosis': [stats.kurtosis(binomial_samples), stats.kurtosis(normal_samples), stats.kurtosis(poisson_samples)]
}
pd_table = pd.DataFrame(data)
print(pd_table)

```
